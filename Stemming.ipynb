#üîπ Stemming in NLP: A Complete Guide
##1Ô∏è‚É£ What is Stemming?
Stemming is the process of reducing a word to its base/root form by removing prefixes and suffixes.
* ‚úÖ It is a rule-based technique that does not consider the meaning of words.
* ‚úÖ The resulting stem may not always be a valid word.

* Running ---->	Run
* Studies ---->	Studi
* Happily ---->	Happi
*Played ---->	Play

## 2Ô∏è‚É£ Why Use Stemming?
Stemming is used to normalize words and reduce vocabulary size in NLP tasks such as:

* Text Classification (Sentiment Analysis, Spam Detection)
* Search Engines (Google, ElasticSearch, Lucene)
* Information Retrieval (Finding similar words)
* Topic Modeling (Grouping similar words)

## 3Ô∏è‚É£ How Stemming Works?
* Stemming removes suffixes and prefixes using predefined rules.
* It does not use a dictionary, so sometimes the output might not be a valid word.

# 4Ô∏è‚É£ Types of Stemming Algorithms

## üîπ 1. Porter Stemmer (Most Common)
* ‚úÖ Uses a set of rules to remove suffixes.
* ‚úÖ Fast but may produce non-dictionary words.

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
words = ["running", "played", "studies", "happily"]
stems = [ps.stem(word) for word in words]
print(stems)
```

###üîπ Limitations:

* "Studies" becomes "Studi" (not a valid word).
* "Happily" becomes "Happili".

## üîπ 2. Snowball Stemmer (Improved Porter Stemmer)
* ‚úÖ Works for multiple languages (English, French, German, etc.).
* ‚úÖ More accurate than Porter Stemmer.

```python
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")
words = ["running", "played", "studies", "happily"]
stems = [ss.stem(word) for word in words]
print(stems)
```

## üîπ 3. Lancaster Stemmer (Aggressive)
* ‚úÖ More aggressive than Porter.
* ‚úÖ Can over-stem words, making them too short.

```python
from nltk.stem import LancasterStemmer

ls = LancasterStemmer()
words = ["running", "played", "studies", "happily"]
stems = [ls.stem(word) for word in words]
print(stems)
```

* üîπ Pros: Works well for some words.
* üîπ Cons: Too aggressive ("happily" ‚Üí "happy", "studies" ‚Üí "study").

##6Ô∏è‚É£ Challenges in Stemming
* üî∏ Over-Stemming: Produces too short words.
* üî∏ Under-Stemming: Fails to reduce similar words.
* üî∏ Not Context-Aware: "Better" does not stem to "Good".
* üî∏ Language-Specific: Each language requires custom rules.
