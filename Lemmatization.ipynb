# üîπ Lemmatization in NLP: A Complete Guide

##1Ô∏è‚É£ What is Lemmatization?
Lemmatization is the process of reducing a word to its base or dictionary form (lemma) while considering the word‚Äôs meaning and grammatical context.

* ‚úÖ Unlike stemming, lemmatization produces actual words.
* ‚úÖ Uses a dictionary lookup to get the correct base form.
* ‚úÖ Requires the part of speech (POS) tag for better accuracy.

##2Ô∏è‚É£ Why Use Lemmatization?
Lemmatization is more accurate than stemming because:

* It considers the meaning and context of words.
* It avoids incorrect word truncation (e.g., "studies" ‚Üí "study", not "studi").
* It ensures valid words for text processing in NLP applications.
##üîπ Applications
* ‚úÖ Search Engines ‚Üí "Running" and "Run" should be treated as the same word.
* ‚úÖ Chatbots ‚Üí Understanding sentence meaning.
* ‚úÖ Sentiment Analysis ‚Üí "Better" should be recognized as "Good".
* ‚úÖ Machine Translation ‚Üí Accurate language processing.

##3Ô∏è‚É£ How Lemmatization Works?
Lemmatization requires two things:
* üîπ The word itself
* üîπ The correct Part of Speech (POS) tag

Without a POS tag, words may not be lemmatized correctly.
* üîπ "Running" ‚Üí "Run" (verb)
* üîπ "Better" ‚Üí "Good" (adjective)

#4Ô∏è‚É£ Implementing Lemmatization in Python
##üîπ Using WordNet Lemmatizer (NLTK)

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

words = ["running", "played", "better", "studies", "happily"]

lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
```

```python
print("Without POS Tag:", lemmatized_words)
```

## üîπ Lemmatization with POS Tags
To improve accuracy, we must specify the correct POS tag.

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')

lemmatizer = WordNetLemmatizer()

words = ["running", "played", "better", "studies", "happily"]

lemmatized_words1 = [lemmatizer.lemmatize(word) for word in words]


# Function to get correct POS tag
def get_wordnet_pos(word):
    from nltk.corpus import wordnet
    from nltk import pos_tag
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)  # Default to noun

# Lemmatization with POS tags
lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in words]
```

```python
print("original words" , words)
print("Without POS Tag:", lemmatized_words1)
print("With POS Tag:", lemmatized_words)

```
