#N-Grams in NLP
N-Grams are a technique used in Natural Language Processing (NLP) to analyze sequences of words. Instead of treating words independently (as in Bag of Words), N-Grams consider contiguous sequences of N words, capturing context and phrase structure.



##What are N-Grams?
An N-Gram is a sequence of N words appearing consecutively in a text. It helps in understanding the relationship between words.

1. Unigram (1-Gram): Single words.

2. Bigram (2-Gram): Two consecutive words.

3. Trigram (3-Gram): Three consecutive words.

Higher-order N-Grams: Sequences of four or more words.

1. Unigrams (1-Gram)	['Natural', 'language', 'processing', 'is', 'amazing']
2. Bigrams (2-Gram)	['Natural language', 'language processing', 'processing is', 'is amazing']
3. Trigrams (3-Gram)	['Natural language processing', 'language processing is', 'processing is amazing']

##Why Use N-Grams?
Captures Context: Helps understand phrases instead of single words.

* Useful in NLP Tasks:

* Text classification

* Sentiment analysis

* Speech recognition

* Machine translation

* Predictive Text & Auto-correction: N-Grams are used in keyboards (e.g., "I am going" suggests "to" as the next word).

* Text Similarity & Search Engines: Helps in finding phrase matches.



# Performing N-Grams in Python

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

## Create a Sample Dataset

```python
data = {'Text': [
    'Natural language processing is amazing',
    'Language models are powerful in NLP',
    'Processing text data is a crucial step in NLP',
    'NLP helps machines understand human language'
]}

df = pd.DataFrame(data)
df
```

# Unigram (1-Gram)

```python
vectorizer_unigram = CountVectorizer(ngram_range=(1,1))
X_unigram = vectorizer_unigram.fit_transform(df['Text'])
unigram_df = pd.DataFrame(X_unigram.toarray(), columns=vectorizer_unigram.get_feature_names_out())


print("vocabulary in the text")
print(vectorizer_unigram.vocabulary_)
print("array representation of the Document")
print(X_unigram.toarray())
print("\nUnigram Representation:")
unigram_df
```

# Bigram (2-Gram)

```python
vectorizer_bigram = CountVectorizer(ngram_range=(2,2))
X_bigram = vectorizer_bigram.fit_transform(df['Text'])
bigram_df = pd.DataFrame(X_bigram.toarray(), columns=vectorizer_bigram.get_feature_names_out())


print("vocabulary in the text")
print(vectorizer_bigram.vocabulary_)
print("array representation of the Document")
print(X_bigram.toarray())
print("\nBigram Representation:")
bigram_df
```

# Trigram (3-Gram)

```python
vectorizer_trigram = CountVectorizer(ngram_range=(3,3))
X_trigram = vectorizer_trigram.fit_transform(df['Text'])
trigram_df = pd.DataFrame(X_trigram.toarray(), columns=vectorizer_trigram.get_feature_names_out())



print("vocabulary in the text")
print(vectorizer_trigram.vocabulary_)
print("array representation of the Document")
print(X_trigram.toarray())
print("\nTrigram Representation:")
trigram_df
```
