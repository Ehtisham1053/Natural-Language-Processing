## Step 2: Function to Expand Abbreviations

```python
import re

def expand_abbreviations(text, abbr_dict):
    """
    Expands abbreviations in a given text.

    Args:
    text (str): The input text.
    abbr_dict (dict): Dictionary of abbreviations.

    Returns:
    str: The expanded text.
    """
    words = text.split()  # Split text into words
    expanded_words = [abbr_dict.get(word.lower(), word) for word in words]  # Replace abbreviations
    return " ".join(expanded_words)  # Join words back into a sentence

# Example text
text = "gn bro! idk why u r not coming. btw, cya asap!"

# Expand abbreviations
expanded_text = expand_abbreviations(text, abbreviation_dict)

print("Original Text:", text)
print("Expanded Text:", expanded_text)
```

## Step 3: Expanding Abbreviations in a Dataset (Pandas)

```python
import pandas as pd

data = {'Raw_Text': [
    "gm! how r u?",
    "idk why u r sad, ttyl!",
    "brb, need to go asap."
]}

df = pd.DataFrame(data)

# Apply function to clean dataset
df['Expanded_Text'] = df['Raw_Text'].apply(lambda x: expand_abbreviations(x, abbreviation_dict))

df
```
